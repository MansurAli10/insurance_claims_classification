```python
import warnings

warnings.filterwarnings('ignore')
```

```python
import pandas as pd
```

```python
in_df=pd.read_csv('insurance_claims.csv')
in_df
```

```python
# convert output (Y,N) to (1,0)

fraud_reported={'Y':1,'N':0}

in_df.fraud_reported=[fraud_reported[item] for item in in_df.fraud_reported]
```

```python
in_df
```

```python
# drop the columns

#na>50

na_df=pd.DataFrame({'columns':in_df.columns,'null_per':in_df.isnull().sum()/in_df.shape[0]*100})
# no null values
```

```python
# non-sense columns

non_sense=[col for col in in_df.columns if in_df[col].nunique==1 or in_df.nunique==in_df.shape[0]]
non_sense   # no non-sense columns
```

# chi2 test

```python
from scipy.stats import chi2_contingency

con_cols=[col for col in in_df.columns if in_df[col].dtype=='int64' or in_df[col].dtype=='float64']
cat_cols=[col for col in in_df.columns if in_df[col].dtype=='object']

sig_val=0.05
chi2_corr_cols=[]
chi2_nocorr_cols=[]

for col in cat_cols:
    contingency_table=pd.crosstab(in_df[col],in_df['fraud_reported'])
    p_value=chi2_contingency(contingency_table)[1]
    if p_value<sig_val:
        chi2_corr_cols.append(col)
    else:
        chi2_nocorr_cols.append(col)
    
print(chi2_corr_cols)
print(chi2_nocorr_cols)
```

# ANOVA test

```python
from scipy.stats import f_oneway

sig_val=0.05
anova_corr_cols=[]
anova_non_corr_cols=[]

con_cols=[col for col in in_df.columns if in_df[col].dtype=='int64' or in_df[col].dtype=='float64']

for col in con_cols:
    y_list=list(in_df[in_df['fraud_reported']==1][col])
    n_list=list(in_df[in_df['fraud_reported']==0][col])
    
    p_value=f_oneway(y_list,n_list)[1]
    
    if p_value<sig_val:
        anova_corr_cols.append(col)
    else:
        anova_non_corr_cols.append(col)
        

print(anova_corr_cols)
print(anova_non_corr_cols)
```

# Drop columns

```python
drop_columns=anova_non_corr_cols+chi2_nocorr_cols

y=in_df['fraud_reported']

drop_columns.append('fraud_reported')

in_df.drop(columns=drop_columns,inplace=True)
```

```python
in_df
```

# Train-test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(in_df,y,train_size=0.8,random_state=42)
```

# fill NA values

```python
con_cols=[col for col in in_df.columns if in_df[col].dtype=='int64' or in_df[col].dtype=='float64']
cat_cols=[col for col in in_df.columns if in_df[col].dtype=='object']

for col in con_cols:
    x_train[col].fillna(x_train[col].mean(),inplace=True)
    x_test[col].fillna(x_train[col].mean(),inplace=True)
    
for col in cat_cols:
    x_train[col].fillna(x_train[col].mode()[0],inplace=True)
    x_test[col].fillna(x_train[col].mode()[0],inplace=True)
    
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
std_scalar=StandardScaler()

for col in con_cols:
    x_train[col]=std_scalar.fit_transform(np.array(x_train[col]).reshape(-1,1))
    x_test[col]=std_scalar.transform(np.array(x_test[col]).reshape(-1,1))
```

# Encoding

```python
oe_train=pd.get_dummies(x_train[cat_cols])
oe_test=pd.get_dummies(x_test[cat_cols])
```

```python
print(oe_train.shape)
print(oe_test.shape)  # columns are same no need to align
```

```python
# merging both scaled and encoded columns

x_train_final=pd.concat([x_train[con_cols],oe_train],axis=1)
x_test_final=pd.concat([x_test[con_cols],oe_test],axis=1)
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg=LogisticRegression()

log_reg.fit(x_train_final,y_train)
```

# LASSO and RIDGE

```python
from sklearn.linear_model import SGDClassifier

lasso=SGDClassifier(penalty='l1', alpha=0.01, l1_ratio=0.15)
ridge=SGDClassifier(penalty='l2', alpha=0.01, l1_ratio=0.15)

lasso.fit(x_train_final,y_train)
ridge.fit(x_train_final,y_train)

lasso_coeffs=lasso.coef_
ridge_coeffs=ridge.coef_


len(lasso_coeffs[lasso_coeffs==0])
len(ridge_coeffs[ridge_coeffs==0])

lasso_test_pred=lasso.predict(x_test_final)
ridge_test_pred=ridge.predict(x_test_final)

from sklearn.metrics import accuracy_score

print('lasso_test_accuracy:',accuracy_score(y_test,lasso_test_pred))
print('ridge_test_accuracy:',accuracy_score(y_test,ridge_test_pred))
```
